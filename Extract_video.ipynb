**Extract không giới hạn số lượng frame**

```python
import cv2
import os


# Path to the video file
input_folder_path = r'C:\Users\Uyen\Documents\PhanTichDLChuyenBiet\CV _ Speech\Data\liveness-20231228T102021Z-001\liveness\train\videos'

# Create a new folder for saving extracted frames
output_folder_path = r'C:\Users\Uyen\Documents\PhanTichDLChuyenBiet\DATA\train'
if not os.path.exists(output_folder_path):
    os.makedirs(output_folder_path)


# List all video files in the input folder
video_files = [f for f in os.listdir(input_folder_path) if f.endswith(('.mp4', '.avi', '.mkv'))]


for video_file in video_files:
    # Create a sub-folder inside the output folder for each video
    video_output_path = os.path.join(output_folder_path, video_file.split('.')[0])
    if not os.path.exists(video_output_path):
        os.makedirs(video_output_path)

    # Open the video file
    video_path = os.path.join(input_folder_path, video_file)
    cap = cv2.VideoCapture(video_path)
    frame_count = 0

    while True:
        ret, frame = cap.read()

        if not ret:
            break

        # Save the frame as an image file
        frame_name = f"frame_{frame_count}.jpg"
        frame_path = os.path.join(video_output_path, frame_name)
        cv2.imwrite(frame_path, frame)

        frame_count += 1

    # Release the video capture object
    cap.release()

print("Frames extraction completed!")
```

**Extract giới hạn số lượng frame**

```python
import os
import cv2

# Path to the video file
input_folder_path = r'C:\Users\Uyen\Documents\PhanTichDLChuyenBiet\CV _ Speech\Data\liveness-20231228T102021Z-001\liveness\train\videos'

# Create a new folder for saving extracted frames
output_folder_path = r'C:\Users\Uyen\Documents\PhanTichDLChuyenBiet\data2\train'
if not os.path.exists(output_folder_path):
    os.makedirs(output_folder_path)

# List all video files in the input folder
video_files = [f for f in os.listdir(input_folder_path) if f.endswith(('.mp4', '.avi', '.mkv'))]

# Maximum number of frames to extract per video
max_frames_per_video = 5

for video_file in video_files:
    # Create a sub-folder inside the output folder for each video
    video_output_path = os.path.join(output_folder_path, video_file.split('.')[0])
    if not os.path.exists(video_output_path):
        os.makedirs(video_output_path)

    # Open the video file
    video_path = os.path.join(input_folder_path, video_file)
    cap = cv2.VideoCapture(video_path)
    frame_count = 0

    while True:
        ret, frame = cap.read()

        if not ret or frame_count >= max_frames_per_video:
            break

        # Save the frame as an image file
        frame_name = f"frame_{frame_count}.jpg"
        frame_path = os.path.join(video_output_path, frame_name)
        cv2.imwrite(frame_path, frame)

        frame_count += 1

    # Release the video capture object
    cap.release()

print("Frames extraction completed!")
```

```python
import os
import cv2

# Path to the video file
input_folder_path = r'C:\Users\Uyen\Documents\PhanTichDLChuyenBiet\CV _ Speech\Data\liveness-20231228T102021Z-001\liveness\train\videos'

# Create a new folder for saving extracted frames
output_folder_path = r'C:\Users\Uyen\Documents\PhanTichDLChuyenBiet\data2\train'
if not os.path.exists(output_folder_path):
    os.makedirs(output_folder_path)

# List all video files in the input folder
video_files = [f for f in os.listdir(input_folder_path) if f.endswith(('.mp4', '.avi', '.mkv'))]

# Number of frames to extract per video
max_frames = 5

for video_file in video_files:
    # Create a sub-folder inside the output folder for each video
    video_output_path = os.path.join(output_folder_path, video_file.split('.')[0])
    if not os.path.exists(video_output_path):
        os.makedirs(video_output_path)

    # Open the video file
    video_path = os.path.join(input_folder_path, video_file)
    cap = cv2.VideoCapture(video_path)

    # Get the total duration of the video
    total_duration = cap.get(cv2.CAP_PROP_FRAME_COUNT) / cap.get(cv2.CAP_PROP_FPS)

    # Calculate the time interval between frames
    time_interval = total_duration / max_frames

    frame_count = 0
    current_time = 0.0

    while True:
        ret, frame = cap.read()

        if not ret:
            break

        # Check if it's time to save a frame
        if cap.get(cv2.CAP_PROP_POS_MSEC) >= current_time * 1000:
            # Save the frame as an image file
            frame_name = f"frame_{frame_count}.jpg"
            frame_path = os.path.join(video_output_path, frame_name)
            cv2.imwrite(frame_path, frame)

            frame_count += 1
            current_time += time_interval

            if frame_count >= max_frames:
                break

    # Release the video capture object
    cap.release()

print("Frames extraction completed!")
```

**Di chuyển hình ảnh vào thư mục test**

```python
import os
import shutil

# Thay thế 'đường_dẫn_thư_mục_gốc' bằng đường dẫn thư mục gốc của bạn
public_test_2_folder = r'C:\Users\Uyen\Documents\PhanTichDLChuyenBiet\data2\public_test_2'

public_test_2_subfolders = [subfolder for subfolder in os.listdir(public_test_2_folder) if os.path.isdir(os.path.join(public_test_2_folder, subfolder))]
for subfolder in public_test_2_subfolders:
    subfolder_path = os.path.join(public_test_2_folder, subfolder)
    for file in os.listdir(subfolder_path):
        file_path = os.path.join(subfolder_path, file)
        destination_path = os.path.join(public_test_2_folder, file)

        # Check if the file already exists in the destination
        counter = 1
        while os.path.exists(destination_path):
            file_name, file_extension = os.path.splitext(file)
            destination_path = os.path.join(public_test_2_folder, f"{file_name}_{counter}{file_extension}")
            counter += 1

        shutil.move(file_path, destination_path)
    os.rmdir(subfolder_path)


print("Image files moved successfully.")
```
